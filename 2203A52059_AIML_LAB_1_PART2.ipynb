Visualization of Data using matplotlib, pyplots Packages

Generate a Data set for Health Care using the following: people = ['kiran', arun', 'vijay', 'varun'] age =[25, 30, 35, 40, 45] height =[145, 151, 165, 173] weight=[45, 55, 65, 75]

```python
import matplotlib.pyplot as plt


people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]  # Assuming an additional age for varun
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
```

Scatter Plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(age,height,color='blue',label='Height')
plt.scatter(age,weight,color='red',label='Weight')
plt.title('Scatter Plot of Age vs. Height and Weight')
plt.xlabel('age')
plt.ylabel('height and weight')
plt.legend()
plt.grid(True)
plt.show()
```

Bar Chart

```python
plt.figure(figsize=(8, 6))
bar_width = 0.35
bar_positions = [i for i in range(len(people))]
plt.bar(bar_positions, height, width=bar_width, color='blue', label='Height')
plt.bar([pos + bar_width for pos in bar_positions], weight, width=bar_width, color='red', label='Weight')
plt.xlabel('People')
plt.ylabel('Height and Weight')
plt.title('Bar Chart of Height and Weight for Each Person')
plt.xticks([pos + bar_width / 2 for pos in bar_positions], people)
plt.legend()
plt.show()
```

Histogram

```python
plt.figure(figsize=(8, 6))
plt.hist(height, bins=20, color='blue', alpha=0.7, label='Height')
plt.hist(weight, bins=20, color='red', alpha=0.7, label='Weight')
plt.title('Histogram of Height and Weight')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
